# Lyric Annotation

### PREVIOUS WORKS

Previous research has explored the problem of automatic lyrics transcription using various machine learning approaches. 
The Whisper model developed by OpenAI is a large-scale speech recognition system trained with weak supervision that performs well on 
general audio transcription tasks, though music introduces additional challenges such as extended vowels and background instrumentation (
Radford et al., 2022). Huang (2025) provides an in-depth analysis of automatic lyrics transcription and highlights 
several key difficulties, including variable tempo, singer style, and acoustic variability. Additionally, the DALI dataset, 
created by Meseguer-Brocal et al. (2018), provides thousands of songs with time-aligned audio, lyrics, and notes, 
enabling research on both lyric and musical transcription tasks. 
These works provide important foundations and resources for further research in automated music transcription systems.


### RESEARCH QUESTIONS TO BE ANSWERED

How does my model perform compared to OpenAI's Whisper?

Does the genre of a song affect how accurately lyrics can be transcribed by the model?
Null Hypothesis (H₀)
    There is no difference in transcription accuracy between music genres.
Alternative Hypothesis (H₁)
    At least one genre has a different transcription accuracy.

### HYPOTHESIS

A machine learning model can be developed to accurately transcribe song audio into lyrics text.

Using MFCC audio representations and a neural network architecture will allow effective learning of lyric patterns from audio signals.

The proposed model can achieve competitive transcription performance, which will be evaluated against an established model (OpenAI Whisper).

The system may also be extendable to transcribe musical notes and chords in addition to lyrics.



### METHODOLOGY

Data Processing

Audio files will be converted into MFCCs (Mel-frequency cepstral coefficients), which represent audio signals numerically over time.

This transformation produces vector representations suitable for machine learning models.

### Model Architecture


A convolutional neural network (CNN) will be built and trained on MFCC representations.

The network will contain multiple hidden layers that will be fine-tuned to optimize performance.

### Model Components


The implementation will include:

Linear projection to manipulate vector inputs for efficiency.

BiLSTM (Bidirectional Long Short-Term Memory) layers for sequence modeling.

LogSoftmax as the classification function.

CTC (Connectionist Temporal Classification) loss to evaluate prediction alignment with transcription outputs.


### Training Objective

The model will take an audio file as input.

The model will output a text transcription of the lyrics.


### Evaluation Method

Model performance will be measured using Word Error Rate (WER).

Results will be compared against OpenAI Whisper.


### TESTING METHODOLOGY

Chi2 test will be used for both research questions.
